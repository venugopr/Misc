```python
# prompt: generate a code for straight line to and plot in matplotlib

import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)

# Define the slope and y-intercept for the line
slope = 2
y_intercept = 1

# Calculate corresponding y values using the equation of a line: y = mx + c
y = slope * x + y_intercept

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Straight Line Plot')

# Display the plot
plt.grid(True)
plt.show()
```

```python
# prompt: generate an quation of y = x1w1 + x2w2 + x3w3 in matplotlib with some set of random values

import matplotlib.pyplot as plt
import numpy as np

# Generate random values for x1, x2, x3, w1, w2, w3
x1 = np.random.rand(100)
x2 = np.random.rand(100)
x3 = np.random.rand(100)
w1 = np.random.rand()
w2 = np.random.rand()
w3 = np.random.rand()

# Calculate y values using the equation: y = x1*w1 + x2*w2 + x3*w3
y = x1 * w1

# Create the plot
plt.plot(x1, y, label='y = x1w1 + x2w2 + x3w3')


# Add labels and title
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Equation Plot')
plt.legend()
# Display the plot
plt.grid(True)
plt.show()
```
